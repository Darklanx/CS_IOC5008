```python
import scipy.io as sio
import h5py
import cv2

from hephaestus.models.keras_rcnn_family.mrcnn.utils import *
from hephaestus.models.keras_rcnn_family.mrcnn import utils
from hephaestus.data.image_utils import rle_encode, rle_decode

############################################################
#  Dataset
############################################################
class DemoDataset(utils.Dataset):
    def load_dataset(self, subset='train'):
        """Load a subset of the monuseg dataset.

        dataset_dir: Root directory of the dataset
        image_ids: List of image ids, you should train/test split by image ids, e.g.
                   image_ids = TRAIN_IMAGE_IDS / image_ids = VAL_IMAGE_IDS
        """
        # Add classes. We have one class.
        # Naming the dataset demo, and the class demo_class1
        self.subset = subset
        if self.subset == "train":
            self.coco = COCO("pascal_train.json")
        else:
            self.coco = COCO("pascal_test.json") 
        
        for i in self.coco.cats:
            self.add_class(source="voc", class_id=i, class_name=self.coco.cats[i]['name'])
  
        # Add images
        for image_id in self.coco.imgs:
            self.add_image(
                "svhn",
                 image_id=image_id,
                 path=os.path.join("{}.jpg".format(image_id))) # define your path of your images
        
#         self.dataset_dir = dataset_dir
        
            
    def load_image(self, image_id):
        # Load image
        if self.subset == 'train':
            print("train_images/" + self.coco.loadImgs(image_id)[0]['file_name'])
            image = cv2.imread("train_images/" + self.coco.loadImgs(image_id)[0]['file_name'])[:,:,::-1] # convert to RGB
        else:
            image = cv2.imread("test_images/" + self.coco.loadImgs(image_id)[0]['file_name'])[:,:,::-1] # convert to RGB
     
            
        # If grayscale. Convert to RGB for consistency.
        if image.ndim != 3:
            image = skimage.color.gray2rgb(image)
        # If has an alpha channel, remove it for consistency
        if image.shape[-1] == 4:
            image = image[..., :3]
        return image
        
    def load_mask(self, image_id):
        """Generate instance masks for an image.
       Returns:
        masks: A bool array of shape [height, width, instance count] with
            one mask per instance.
        class_ids: a 1D array of class IDs of the instance masks.
        """
        
        annids = self.coco.getAnnIds(imgIds=image_id)
        anns = self.coco.loadAnns(annids)
        
        mask_list = []
        id_list = []
        for i in range(len(annids)):
            mask = self.coco.annToMask(anns[i])
            cate = self.coco.loadAnns(annids[i])[0]['category_id']   
            
            mask_list.append(mask)
            id_list.append(cate)
            
        mask = np.stack(mask, axis=-1)
        return mask, np.array(id_list).astype(np.int32)
 
    def read_data_and_mask_arr(self):
        """Read all data into memory to speed up training
        If you want to pre-load data into memory, set --load_data_to_ram=True when running train.py
        """
   
        print("Pre-loading data into memory")
        self.image_arr = np.array([self.load_image(x) for x in tqdm(self.image_ids)])
        
        # Since maks has different depth in each image, we store mask arr in dict
        self.mask_arr = {}
        for x in tqdm(self.image_ids):
            self.mask_arr[x] = self.load_mask(x)
        
    def load_mask_arr(self, image_id):
        """Get instance masks from image array
        Returns:
        masks: A bool array of shape [height, width, instance count] with
            one mask per instance.
        class_ids: a 1D array of class IDs of the instance masks.
        """
        mask, class_ids = self.mask_arr[image_id]
        return mask, class_ids

    def image_reference(self, image_id):
        """Return the path of the image."""
        info = self.image_info[image_id]
        if info["source"] == "nuclear":
            return info["id"]
        else:
            super(self.__class__, self).image_reference(image_id)



from hephaestus.models.keras_rcnn_family.yacs_config import *
from hephaestus.models.keras_rcnn_family.mrcnn.model import *
from hephaestus.models.keras_rcnn_family.mrcnn.utils import *

config = get_cfg_defaults()
config.merge_from_file("config.yaml")
# config.GPU_COUNT = 1
# config.IMAGES_PER_GPU = 1
config = parse_config(config)
print(config)
```
